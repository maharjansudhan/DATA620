```python
# importing required packages for this project
import networkx as nx
import networkx.generators.small
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# let's create a graph in which we can work on. Since we have not put any nodes yet there is an empty graph 
Graph = nx.generators.small.krackhardt_kite_graph()
```

```python
# Since we know that there are 10 people, according to the krackhardt_kite_graph there will be 18 edges for 10 nodes.
# from A to J, we have 10 names

nodes = ['Andre','Beverly', 'Carol', 'Diane', 'Ed', 'Fernando','Garth','Heather','Ike', 'Jane']
Graph.add_nodes_from(nodes)
```

```python
# now let's add the edges, there are 18 connections
edges = [('Andre', 'Beverly'), ('Andre', 'Carol'), ('Andre', 'Diane'), ('Andre', 'Fernando'),
         ('Beverly', 'Diane'), ('Beverly', 'Ed'), ('Beverly', 'Garth'),
         ('Carol', 'Diane'), ('Carol', 'Fernando'),
         ('Diane', 'Ed'), ('Diane', 'Fernando'), ('Diane', 'Garth'),
         ('Ed', 'Garth'),
         ('Fernando', 'Garth'),
         ('Fernando', 'Heather'),
         ('Garth', 'Heather'),
         ('Heather', 'Ike'),
         ('Ike', 'Jane'),]
Graph.add_edges_from(edges)
```

```python
# now let's plot the graph


plt.figure(figsize=(8, 6))
nx.draw(Graph, node_size=20, node_color='black')
plt.title('Hello, Graph World', size=15)
plt.show()
```
